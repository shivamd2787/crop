```python
from sklearn.preprocessing import LabelEncoder 
```

```python
lb=LabelEncoder()
```

```python
df['crop_encode']=lb.fit_transform(df['label'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['label','crop_encode'],axis=1)
```

```python
x
```

```python
y=df['crop_encode']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2529)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)
```

```python
import pickle    #file handelling technique
```

```python
with open('model.pkl','wb') as model_file:
    pickle.dump(lg,model_file)  #here lg means logistic regression
```
